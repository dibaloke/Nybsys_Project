```python
#Import Libraries 

import pixellib
import cv2
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pixellib.instance import instance_segmentation
```

```python
# Segment Images 

seg = instance_segmentation()
seg.load_model("mask_rcnn_coco.h5")
segmask, output = seg.segmentImage("sample.jpeg", show_bboxes=True, output_image_name="output.jpg",
extract_segmented_objects= True, save_extracted_objects=False)
```

```python
# Load ms_coco label csv

df=pd.read_csv('coco_labels.csv')
df
```

```python
# Segmented Objects' Class IDs

labels = list(segmask['class_ids'])
labels
```

```python
# Get Class Names from the Class IDs

class_names=[]

for label in labels:
    filt=(df['Class_id']==label)
    class_names.append(df.loc[filt,'Class_Name'].values[0])
```

```python
# Create Dictionary for count of objects per class

class_count_dict = dict((i, labels.count(i)) for i in labels)
class_count_dict
```

```python
# Separately save the segmented objects with their corresponding class name along with count 
# Create a lookup table so we can refer to the exact mask based on name 

num_objects = len(segmask['class_ids'])
lookup_table=[]

for i in range(0,num_objects):
    current_label = labels[i]
    cv2.imwrite(os.path.join('./',class_names[i]+str(class_count_dict[current_label])+'.jpg'),segmask['extracted_objects'][i])
    lookup_table.append(class_names[i]+str(class_count_dict[current_label])+'.jpg')
    class_count_dict[current_label]= class_count_dict[current_label]-1
```

```python
lookup_table
```

```python
# Take User Input 

target_object = input('Please input the name of the object you want to keep unblurred:')
```

```python
target_object = target_object+'.jpg'
```

```python
# Check if the lookup table works (ignore the color map !!  This is just for sanity check)

plt.imshow(segmask['extracted_objects'][lookup_table.index(target_object)])
```
